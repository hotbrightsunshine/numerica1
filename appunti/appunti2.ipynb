# Appunti seconda lezione di Matplotlib

## Curve isoparametriche

Vogliamo visualizzare funzioni della forma 

$$
 z = f(x,y)
$$

Fissato $N \in \mathbb{N}$, otteniamo $\{L_i\}_{i = 0}^{N}$ $N$ linee di livello per la funzione $z$. Coloriamo secondo una scala di colorazione il piano cartesiano separato dalle $L_i$. 

```python
t = np.arange(-2, 2, 0.01)
X, Y = np.meshgrid(t, t)

Z = np.sin(X * np.pi / 2) + np.cos(Y * np.pi / 4)

plt.figure(figsize=(12, 6))
N = 20

p1 = plt.contourf(X, Y, Z, N, cmap = "jet") # colore
p2 = plt.contour(X, Y, Z, N, colors = ["black"]) # colors è "ciclica" // contorno

plt.clabel(p2, fontsize=15)

plt.colorbar(p1)
```
