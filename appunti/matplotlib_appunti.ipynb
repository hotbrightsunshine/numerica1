# Oggi lezione bellissima (è nero)

Utilizzeremo Myplotlib, ed è un pacchetto che deve essere importato. 


````python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
````

````python
x = np.linspace(0, 2*np.pi, 100)
cosx, sinx = np.cos(x), np.sin(x)

plt.plot(x, cosx)
plt.plot(x, sinx)
````

Le opzioni principali sono 
- `figsize` larghezza e altezza in pollici
- `dpi`
- `facecolor` colore del background
- `linewidth` spessore del bordo esterno dell'intera figura
- `edgecolor` colore del bordo esterno della figura

````python
plt.figure(figsize= (8, 2), facecolor="silver", edgecolor="r", linewidth=2)
plt.plot(x, cosx)
````

## Limiti degli assi e scala

````python
#Range asse X
plt.xlim(-10, 10)

#Range asse Y
plt.xlim(-3, 0)

#oppure, 
plt.axis((-10, 10, -3, 0))
plt.plot(x, cosx)
````

Possiamo imporre l'aspetto del grafico in modo che la distanza tra le tacche di x e y sia uguale. 

````python
plt.gca().set_aspect("equal")
plt.plot(x, cosx)
````

## Tacchette dei grafici

````python
plt.figure(figsize=(10, 3))
plt.xticks(np.linspace(np.min(x), np.max(x), 10), fontsize=15)
plt.plot(x, sinx)
````

# Label 

````python
plt.xlabel("ASSE ICS", fontsize=20)
plt.ylabel("ASSE IPSILON", fontsize = 15)
plt.plot(x, cosx)
````

# Superfici con bordo rettangolare

````python
x = np.linspace(-5, 5, 20)
y = np.linspace(-5, 5, 20)

X, Y = np.meshgrid(x, y) # Griglia

R = np.sqrt(X**2, Y**2) # appoggio
Z = np.sin(R) # Grafico

plt.figure(figsize=(10, 10))

plt.suptitle('Plot della superficie')

ax = plt.subplot(1, 2, 1, projection="3d")
ax.plot_wireframe(X, Y, Z)
ax.set_title("WIREFRAME")

ax = plt.subplot(1, 2, 2, projection="3d")
ax.plot_surface(X, Y, Z, cmap="hot")
````
